```python
import yfinance as yf
import random

def calculate_tri(ftf=0.8, py=0.12, ac=0.3, ahh=24):
    return round(((ftf + py) * (1 + ac)) / ahh, 4)

def rng_keno():
    return random.sample(range(1, 81), 5)

price = yf.download("BTC-USD", period="1d")["Close"].iloc[-1]
tri = calculate_tri()
keno = rng_keno()

print(f"TRI v1.2 SIMULATION LOOP")
print(f"BTC Price (real): ${price.item():.2f}")
print(f"TRI Score: {tri:.4f}")
print(f"RNG Keno draw: {keno}")
print("Loop complete — system live & compounding")
```
